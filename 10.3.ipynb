```python
#importing data
import pandas as pd
import numpy as np
v= pd.read_csv("https://raw.githubusercontent.com/nickeubank/practicaldatascience/master/Example_Data/ca/ca_arrests_2009.csv")
p =pd.read_csv("https://raw.githubusercontent.com/nickeubank/practicaldatascience/master/Example_Data/ca/nhgis_county_populations.csv")
```

```python
m1 = pd.merge(v,p,left_on="COUNTY",right_on="COUNTY",validate="1:m",indicator = True)
m1
```

```python
#7
m2 = pd.merge(v,p,left_on="COUNTY",right_on="COUNTY",validate="1:m",indicator = True,how="left")
m2
```

```python
#8
p2= p1[p1.STATE=="California"]
p1 = p[p.YEAR=="2005-2009"]
p2
```

```python
#8
m3 = pd.merge(v,p2,left_on="COUNTY",right_on="COUNTY",validate="1:1",indicator = True,how="left")
m3._merge.value_counts()
m3[m3._merge !="both"]
```

```python
#10
violent_arrest_rate_2009 = m3['VIOLENT']/m3['total_population']
violent_arrest_rate_2009
```

```python
drug_offenses_2009 = m3['F_DRUGOFF']/m3['total_population']
drug_offenses_2009
```
